```python
from sklearn.model_selection import train_test_split

def prepare_data(X,y,test_size = 0.1):
  index = pd.isnull(X).any(axis=1)
  X = X[~index]
  y = y[~index]
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)
  return X_train, X_test, y_train, y_test
```

```python
result_table =[]
def train_data(X,y,data):
    print(f'-----------{data}----------')
    model = LogisticRegression(class_weight={0:0.3, 1:0.7}, max_iter=10000)
    model.fit(X_train, y_train)

    print('Train size: ', len(X_train))
    train_score = model.score(X_train, y_train)
    print('Training accuracy', train_score)

    #training accuracy
    y_pred = model.predict(X_train)
    cm_train = confusion_matrix(y_train, y_pred)
        # tn, fp, fn, tp = cm.ravel()
    tn, _, _, tp = cm_train.ravel()

    # true positive rate When it's actually yes, how often does it predict yes?
    recall = float(tp) / np.sum(cm_train, axis=1)[1]
    # Specificity: When it's actually no, how often does it predict no?
    specificity = float(tn) / np.sum(cm_train, axis=1)[0]

    print('Train - Recall/ Like accuracy', recall)
    print('Train - specificity/ Dislike accuracy', specificity)
    print()

    #test accuracy
    print('Test accuracy', np.mean(y_test == model.predict(X_test)))
    # report=classification_report(y_test, model.predict(X_test),target_names=['0','1'])
    y_pred = model.predict(X_test)
    cm_test = confusion_matrix(y_test, y_pred)
    tn, fp, fn, tp = cm_test.ravel()

    # true positive rate When it's actually yes, how often does it predict yes?
    recall = float(tp) / np.sum(cm_test, axis=1)[1]
    # Specificity: When it's actually no, how often does it predict no?
    specificity = float(tn) / np.sum(cm_test, axis=1)[0]

    #for plot
    fpr, tpr, _ = roc_curve(y_test, model.predict_proba(X_test)[:,1])
    auc = roc_auc_score(y_test, model.predict_proba(X_test)[:,1])
    result_table.append({'data': data, 'fpr':fpr, 'tpr':tpr, 'auc':auc})


    print('Test - Recall/ Like accuracy', recall)
    print('Test - specificity/ Dislike accuracy', specificity)
    print()

    # # save the model
    # joblib.dump(model, 'log_reg_model.pkl')
    print('auc-train: ', roc_auc_score(y_train, model.predict_proba(X_train)[:,1]))
    print('auc-test: ', auc)
```

```python
def plot_roc():
  fig = plt.figure(figsize=(8,6))

  for model in result_table:
      plt.plot(model['fpr'], 
              model['tpr'], 
              label= f"{model['data']}, AUC={round(model['auc'],3)}" )
      
  plt.plot([0,1], [0,1], color='orange', linestyle='--')

  plt.xticks(np.arange(0.0, 1.1, step=0.1))
  plt.xlabel("False Positive Rate", fontsize=15)

  plt.yticks(np.arange(0.0, 1.1, step=0.1))
  plt.ylabel("True Positive Rate", fontsize=15)

  plt.title('ROC Curve Analysis', fontweight='bold', fontsize=15)
  plt.legend(prop={'size':13}, loc='lower right')

  plt.show()
```
